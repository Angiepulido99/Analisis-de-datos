```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
Datos = pd.read_csv("C:\Datos2\winequality-red.csv")
```

```python
predictors_col = ["density", "alcohol", "pH",]
target_col = ['quality']
```

```python
predictors = Datos[predictors_col]
target = Datos[target_col]
```

```python
predictors
```

```python
target
```

```python
corr = predictors.corr()
corr
```

```python
sns.heatmap(corr)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.25, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
